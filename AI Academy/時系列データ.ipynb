# 時系列データとは
時系列データとは、時間の変化に注目したデータのことです。
時系列データ解析の例として、株価が上がっているか下がっているかといった時間的な変動を持つ金融データや、気温や雨量といった気象データなどが挙げられます。  
このような時系列データから時間的にどう変化しているのかを解析し、その結果から将来どのように変化するのかを予測することを時系列データ解析といいます。


## 時系列データ解析の一連の流れ
時系列データ解析の一連の流れは次の様に考えることができます。
1. 自己相関と偏自己相関を見る  
↓
2. 成分分解を行なって、トレンド、季節性を抽出  
↓
3. 対数をとって、こちらも同様にトレンド、季節性変動などに分解  
↓
4. 定常性の確認（対数化後）   
↓
5. 出力 p-value 帰無仮説（定常性を持つ）p>0.05が棄却されなければ、定常性を持つとは言えない  
↓
6. 定常性がないらしいので、トレンド和分を差分をとって消す、季節性を12ヶ月差分とって消す  
↓
7. 季節調整をしてから差分を取る  
↓
8. データを学習データとテストデータに分ける  
↓
9. ACFとPACFよりパラメータ候補を選定、季節調整済みの差分データのコレログラム  
↓
10. SARIMA適用   
（SARIMA：Seasonal AutoRegressive Integrated Moving Average/季節自己回帰和分移動平均とは季節S、自己回帰AR、和分I、移動平均MAが合わさった形）  
sm.tsa.SARIMAX()で利用可能  
↓
11. 予測  
↓
12. 予測の評価  
↓
13. 結果のプロット  

## 時系列データ解析を理解する上で重要な用語  

### AR
AR(AutoRegressive : 自己回帰)モデルは、ある時刻tのデータを、過去の時刻t-kのデータを用いて回帰するモデルです。
### MA
MA(moving average : 移動平均)モデルは、ある時刻のデータが、過去の時刻での誤差項を用いて表現することができます。
### ARMA(自己回帰移動平均)モデル
ARMA(Autoregressive integrated moving average、自己回帰和分移動平均)モデルとは、AR(Auto Regressive/自己回帰モデル)モデルとMA(Moving Average/移動平均モデル)モデルから構成されるモデルであり、ある時点の出力が、過去の出力と、現在および過去の入力に対する和で表現されるモデルです。
### ARIMAモデル
ARIMA（Autoregressive Integrated and Moving Average:自己回帰和分移動平均）過程モデルとは、ある時系列データに対して何度か差分を取った後に、データがARMAモデルで表現することが可能です。


## 自己相関と偏自己相関
**自己相関（ACF: Autocorrelation Function）は、ある過去の値と現在のデータにどれくらい影響しているか（似ているか）を表したものです。**  
<font color="Crimson">**過去の値が現在のデータにどれくらい影響しているかの関係性を調べる際に自己相関を利用します。**</font>
自己相関係数の値の範囲は-1~1をとり、正負の関係や絶対値が1に近いほど相関が強いと考えられます。
自己相関はその名の通り自分自身との相関ですので、「元のデータ」と「元のデータから時間をずらしたデータ」を利用して相関係数を計算します。
この時、正の自己相関があれば、先月の乗客数が多ければ、今月も多く、負の自己相関であれば、先月の乗客数が少なければ、今月も少ないということを表します。
例えば、今日と前日が似ている場合、前日と一昨日が似ているといったように考えられますが、一昨日と今日がは似ているのか似ていないのかといった判別が難しいため、このような場合には「偏自己相関」を用いることで、ある特定の日のみと自己相関を見ることが出来ます。
Pythonでは、自己相関をtsa.stattools.acf()で求められます。
**偏自己相関（PACF: Pertial Autocorrelation Function）は、自己相関係数から、時間によって受ける影響を無視（除去）して計算された自己相関です。**
偏自己相関を使うと、例えば1日前の影響を除き今日と二日前だけの関係を調べたいといった場合に利用します。
Pythonでは偏自己相関をtsa.stattools.pacf()で求められます。


## トレンド、周期変動、不規則変動
**時系列データにはトレンド、周期変動、不規則変動（その他の残差）の3つのパターンがあると考えられます。
トレンドはデータの長期的な傾向を意味します。**  
時間の経過とともにデータの値が上昇していたり下降していたりする時系列データは「トレンドがある」といい、値が上昇している場合には 正のトレンド 下降している場合には 負のトレンド があると言います。  
**周期変動 があるデータは時間の経過に伴ってデータの値が上昇と下降を繰り返します。特に1年間での周期変動を 季節変動 といいます。**  
**不規則変動 は時間の経過と関係なくデータの値が変動することをいいます。**  
実際の時系列データを観測すると、これら3つのパターンが組み合わさっていることが分かります。

### データを作成する
データを作成するには、pandasのdate_range()が便利です。

```
import pandas as pd

# 1ヶ月分のサンプルデータは次のように作れます。
d = pd.date_range(start="2020-12-01", end="2020-12-31")
print(d)
```

## 日経経平均株価の時系列データ解析
StatsModelsと使って日経平均株価のデータを分析します。
まずは以下のURLから日経平均株価の月次データをダウンロードしてください。
日経平均株価の月次データ  
https://indexes.nikkei.co.jp/nkave/index?type=download  

image.png

```
import numpy as np
import matplotlib.pyplot as plt
# from pandas.core import datetools
import statsmodels.api as sm

df = pd.read_csv('nikkei_stock_average_monthly_jp.csv', encoding='shift_jis')
df = df.iloc[:-1, 0:2] # データ日付と終値だけを取得(始値や安値などは除く)
df.head() # 先頭から5件取得
```

データの可視化をしてみましょう。

```
"""

date = pd.to_datetime(date) の行部分にて、
環境によってはワーニングエラーが出る場合があります。

その場合、下記の２行を先頭に追加し、実行して見てください。

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

"""

df.columns = ['date', 'price']

date = np.array(df['date'])
price = np.array(df['price'])

date = pd.to_datetime(date) # この行で環境によってはワーニングエラーが出る場合があります。

plt.figure(figsize=[20, 5]) # widthが20でheightが5
plt.style.use('ggplot')

plt.plot(date, price)
plt.title('Nikkei Price') # titleにNikkei Priceとつける
plt.xlabel('date') # x軸のラベルにdateとつける
plt.ylabel('price')# y軸のラベルにpriceとつける

df['date'].head()
```

自己相関と偏自己相関をプロットします。  
また、元データから時間をずらした量(ラグ)と自己相関を表したグラフをコレログラムと言います。

```
price = df.set_index('date')

# 自己相関
price_acf = sm.tsa.stattools.acf(price, nlags=40)
# 偏自己相関
price_pacf = sm.tsa.stattools.pacf(price, nlags=40, method='ols')

#  自己相関のグラフ（コレログラム）
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(211)

# ラグが40(lags=40)までの自己相関
fig = sm.graphics.tsa.plot_acf(price, lags=40, ax=ax1)
ax2 = fig.add_subplot(212)

# ラグが40(lags=40)までの編自己相関
fig = sm.graphics.tsa.plot_pacf(price, lags=40, ax=ax2)
```

次に、トレンド、周期変動、残差に分解します。
StatsModelsでは、statsmodels.tsa.api.seasonal_decomposeを使うことで分解できます。  

下記グラフは、上から元データ、トレンド、周期変動（季節変動,Sesonal）、残差の順に並んでいます。

```
#  グラフのサイズを変更
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 20,20

seasonal = sm.tsa.seasonal_decompose(price, freq=15)
seasonal.plot()
```

### 予測モデルの作成
それでは予測モデルを作っていきます。
ProphetではProphet()をインスタンス化します。
その後、.fit()にて学習を行います。
引数には、先ほど読み込んだデータフレームdfを渡します。

```
m = Prophet()
m.fit(df)
```

### 予測した期間を追加したデータフレームを作成
次に予測を行うために、新規に予測用のデータを作成します。  
予測を行うには、日付を含むdsカラムが入ったデータフレームを利用する必要があります。  
**make_future_dataframeメソッドを使うことで、予測した期間を追加したデータフレームを作成出来ます。**

```
future = m.make_future_dataframe(periods=365)
future.tail()
```

### 予測と可視化
それでは先ほど作成したfutureを用いて、predictメソッドにて予測を行います。
ここでカラムに関して説明致します。  
yhatは予測値になります。  
次にyhat_lowerとyhat_upperですが、それぞれ区間予測の上下限値になります。
futureの各行をyhatという予測値にあてはめます。学習させた時系列データより先の日については、学習データを用いて予測値を算出します。

```
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```
fig1 = m.plot(forecast)
```

予測の構成要素を確認するにはplot_componentsメソッドを使います。
初期設定ではトレンドに加え、時系列データの年・週単位の周期性が表示されます。

```
fig2 = m.plot_components(forecast)
```
